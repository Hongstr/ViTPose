```python
jsonfile = "result_keypoints.json"
annotationfile = "/home/shanchuan/data/animal_dataset_v1_c/japanesemacaque_test.json"
```

```python
import json
```

```python
pred_file = None
with open(jsonfile, 'r') as f:
    pred_file = json.load(f)
gt_file = None
with open(annotationfile, 'r') as f:
    gt_file = json.load(f)
```

```python
print(pred_file[0])
print(gt_file['annotations'])
```

```python
gt = {}
for i in gt_file['annotations']:
    gt[i['image_id']] = i
```

```python
def format_keypoints(raw_keypoints):
    keypoints = []
    for i in range(0, len(raw_keypoints), 3):
        keypoints.append(raw_keypoints[i:i+2])
    return keypoints
# format_keypoints(pred_file[0]['keypoints'])
# format_keypoints(gt[101]['keypoints'])
```

```python
import numpy as np
import math
def compute_nme(pred_pts, gt_pts, area):
    nme = 0
    for i in range(len(pred_pts)):
        nme += np.linalg.norm(np.array(pred_pts[i]) - np.array(gt_pts[i]))
    return nme / (len(pred_pts) * math.sqrt(area))
```

```python
rmse_sum = 0
for i in pred_file:
    pred_keypoints = format_keypoints(i['keypoints'])
    gt_keypoints = format_keypoints(gt[i['image_id']]['keypoints'])
    area = gt[i['image_id']]['area']
    rmse = compute_nme(pred_keypoints, gt_keypoints, area)
    rmse_sum += rmse
    print(rmse)
print("nme: " + str(rmse_sum / len(pred_file)))
```
